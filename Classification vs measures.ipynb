# Load all the data.

Columns are: classification,measure_new_score,measure_coursestddev,measure_speedddev,measure_speeddavg

```python
x = np.loadtxt("classification_vs_measures.csv", delimiter=",", skiprows=1)
```

```python
histfig = figure(figsize=(20,5))
subplot = histfig.add_subplot(111)
fishy = subplot.hist(x[x[:,0] > 0.5][:,1], bins=200, normed=False, color='b', alpha=0.5, label="fishy score")
nonfishy = subplot.hist(x[x[:,0] < 0.5][:,1], bins=200, normed=False, color='r', alpha=0.5, label="nonfishy score")
legend()
show()
```

```python
histfig = figure(figsize=(20,5))
subplot = histfig.add_subplot(121)
subplot.hist(x[x[:,0] > 0.5][:,2], bins=200, normed=True, color='b', alpha=0.5, label="fishy course stddev")
subplot.hist(x[x[:,0] < 0.5][:,2], bins=200, normed=True, color='r', alpha=0.5, label="nonfishy course stddev")
legend()

subplot = histfig.add_subplot(122)
subplot.hist(log10(x[x[:,0] > 0.5][:,2]+0.001), bins=200, normed=True, color='b', alpha=0.5, label="fishy course stddev log")
subplot.hist(log10(x[x[:,0] < 0.5][:,2]+0.001), bins=200, normed=True, color='r', alpha=0.5, label="nonfishy course stddev log")
legend()
show()
```

```python
histfig = figure(figsize=(20,5))
subplot = histfig.add_subplot(121)
subplot.hist(x[x[:,0] > 0.5][:,3], bins=200, normed=True, color='b', alpha=0.5, label="fishy speed stddev")
subplot.hist(x[x[:,0] < 0.5][:,3], bins=200, normed=True, color='r', alpha=0.5, label="nonfishy speed stddev")
legend()

subplot = histfig.add_subplot(122)
subplot.hist(log10(x[x[:,0] > 0.5][:,3]+0.001), bins=200, normed=True, color='b', alpha=0.5, label="fishy speed stddev log")
subplot.hist(log10(x[x[:,0] < 0.5][:,3]+0.001), bins=200, normed=True, color='r', alpha=0.5, label="nonfishy speed stddev log")
legend()
show()
```

```python
histfig = figure(figsize=(20,5))
subplot = histfig.add_subplot(111)
fishy = subplot.hist((x[x[:,0] > 0.5][:,4]+0.001), bins=200, normed=True, color='b', alpha=0.5, label="fishy speed avg")
nonfishy = subplot.hist((x[x[:,0] < 0.5][:,4]+0.001), bins=200, normed=True, color='r', alpha=0.5, label="nonfishy speed avg")
legend()
show()
```

```python
histfig = figure(figsize=(20,10))
subplot = histfig.add_subplot(211)
subplot.hist((x[x[:,0] > 0.5][:,1]), bins=200, normed=False, color='b', alpha=0.5, label="fishy score")
subplot.hist((x[x[:,0] < 0.5][:,1]), bins=200, normed=False, color='r', alpha=0.5, label="nonfishy score")
legend()

subplot = histfig.add_subplot(212)
xfishy = x[x[:,0] > 0.5][:,4]
xfishy = where(xfishy > 1.0, 1.0, where(xfishy < 0.0, 0, xfishy))
xnonfishy = x[x[:,0] < 0.5][:,4]
xnonfishy = where(xnonfishy > 1.0, 1.0, where(xnonfishy < 0.0, 0, xnonfishy))
subplot.hist(xfishy, bins=200, normed=False, color='b', alpha=0.5, label="fishy speed avg")
subplot.hist(xnonfishy, bins=200, normed=False, color='r', alpha=0.5, label="nonfishy speed avg")
legend()

show()
```
